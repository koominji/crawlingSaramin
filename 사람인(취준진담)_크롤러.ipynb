```python
#필요한 라이브러리 추가

from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# 사람인 -> 취준진담 -> 취업준비

driver_path='chromedriver.exe'

driver = webdriver.Chrome(executable_path=driver_path)
driver.implicitly_wait(3)
driver.get("https://www.saramin.co.kr/zf_user/company-review-qst-and-ans/sub?page=1&prev=sub&keyword=%EC%B7%A8%EC%97%85%EC%A4%80%EB%B9%84&csn=&job_cd=&sort=reg_dt&isAnswerable=n&myQustViewFl=n&influencerFl=n&searchType=hashtag&banner_idx=&influencer=")
driver.implicitly_wait(3)

time.sleep(0.5)
```

```python
result=[]
resultList=[]
```

```python
result=[]
resultList=[]

pageNum = 1

# 취준진담 마지막 페이지인 476페이지까지 크롤링 시도
while pageNum<476:
    time.sleep(2)
    driver.get(f"https://www.saramin.co.kr/zf_user/company-review-qst-and-ans/sub?page={pageNum}&prev=sub&keyword=취업준비&csn=&job_cd=&sort=reg_dt&isAnswerable=n&myQustViewFl=n&influencerFl=n&searchType=hashtag&banner_idx=&influencer=")
    pageNum += 1
    
    page = driver.page_source
    soup = BeautifulSoup(page,"html.parser") #html로 파싱
    
    # 각페이지에 있는 글을 가져온다
    each_article=soup.find_all('a',{'class':'link_detail'})
   
    
    for article in each_article:
        each_article_page='https://www.saramin.co.kr'+article['href']
        driver.get(each_article_page)
        
        each_page_source= driver.page_source
        eachArticle_info_soup = BeautifulSoup(each_page_source,"html.parser")
        
        ## 제목 title
        title = eachArticle_info_soup.find('h1',{'class':'tit_subject'}).text
        
        ## 내용+댓글 result 
        result=""
        contents=eachArticle_info_soup.find_all('p',{'class':'txt_desc'})
        for c in contents:
            result += c.text
        
        print(result)
        result=[title,result]
        resultList.append(result)

        
    result_df = pd.DataFrame(resultList,columns=['title','content'])
```
